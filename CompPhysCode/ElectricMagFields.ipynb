# Making it Round the Bend 

```python
import numpy as np 
import matplotlib.pyplot as plt 
```

## Part 1 

```python
# Add our experimental data from ion accelerator 
Efield = np.array([1000,1600,2200,3000,4000]) #N/C 
a = np.array([1.0157e11,1.6233e11,2.2452e11,3.0616e11,4.0815e11]) #m/s/s 
```

```python
# Identify linear regression line 
m, b = np.polyfit(a, Efield, 1)
print('y = ', round(m,10), 'a +', round(b,2))
```

```python
# Visualize 
plt.plot(a, m*a+b, label = 'y =  9.8e-9 x + 8.26', color = 'g') 
plt.scatter(a, Efield, color = 'm', label = 'Experimental data', marker = '*')
plt.title("Mass to Charge Ratio")
plt.xlabel("Acceleration (m/s/s)") 
plt.ylabel("Electric Field Strength (N/C)") 
plt.legend()
```

## Part 2 

```python
# Add data 
E = np.array([460, 1000, 1200, 1500, 2000]) #N/C 
B = np.array([0.005, 0.01, 0.012, 0.015, 0.02]) #T 

# Identify linear regression line 
m, b = np.polyfit(B, E, 1)
print('y = ', round(m,10), 'a +', round(b,2)) 

# Visualize 
plt.plot(B, m*B+b, color = 'c', label = 'y =  1e5 x + -37.32') 
plt.scatter(B, E, color = 'b', label = 'Experimental data', marker = 'd')
plt.title("Velocity of the Charged Particle")
plt.xlabel("Magnetic Field Strength (T)") 
plt.ylabel("Electric Field Strength (N/C)") 
plt.legend()
```

## Part 3 

```python
r = np.array([-0.0395, -0.0785, -0.028, -0.0215, -0.0195]) # m 
B = np.array([0.05, 0.025, 0.07, 0.09, 0.10]) # T 

# Identify regression polyfit 
m, b = np.polyfit(B, r, 1)
print('y = ', round(m,10), 'a +', round(b,2)) 

# Visualize 
plt.scatter(B, r, color = 'r', label = 'Experimental data')
plt.title("Ratio of Radius to Magnetic Field Strength")
plt.xlabel("Magnetic Field Strength (T)") 
plt.ylabel("Radius of Displacement (m)")  
```

```python
r = np.array([-0.0395, -0.0785, -0.028, -0.0215, -0.0195]) # m 
B = np.array([0.05, 0.025, 0.07, 0.09, 0.10]) # T 

# Identify regression polyfit 
m, b = np.polyfit(1/B, r, 1)
print('y = ', round(m,5), 'a +', round(b,2)) 

# Visualize 
plt.plot(1/B, m*(1/B)+b, label = 'y =  -0.00197 a + 0.0', color = 'b') 
plt.scatter(1/B, r, color = 'r', label = 'Experimental data')
plt.title("Coefficient of Curvature")
plt.xlabel("Inverse of Magnetic Field Strength (1/T)") 
plt.ylabel("Radius of Displacement (m)") 
plt.legend() 
```
