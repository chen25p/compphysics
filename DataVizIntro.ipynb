```python
def model_function(x): # Defines a function 
    a = 2 # Assigns integer 2 to variable a 
    b = -7 # Assigns integer -7 to variable b 
    y = a * (x**2) + b # Creates y variable defined as 'a * x^{2} + b' 
    return y # Outputs the resulting y value based on predetermined constants 

xvals = np.linspace(0, 5, num=10) # A linspace command is used to create an array of 10 even values from 0 to 5 
# Documentation page: https://numpy.org/doc/2.1/reference/generated/numpy.linspace.html 
yvals = model_function(xvals) # The array values (x values) are fed into the model_function and stored as its own array 
plt.scatter(xvals, yvals) # The scatter command is used to generate a scatterplot 
# Documentation page: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html 
```

Below is another line of code that uses the same data but plots differently. Run the code and determine what the difference is.

```python
plt.plot(xvals, yvals, color='k', alpha=.1) # Original code line 
plt.plot(xvals, yvals, color='c', alpha=.1) # Changed the color to cerulean 
plt.plot(xvals, yvals, color='c', alpha=.5) # Raised the alpha (transparency) 
plt.plot(xvals, yvals, color='r', alpha=.5, marker = "$♥$") # Changing the marker shape of the points and color to red (for Valentine's Day) 
```

What happened? How is this different from the previous plot?

#### Next step: Customization

The previous two graphs use the pyplot interface. This interface can generate quick and easy plots, but is best for simple plots.


The code below is an example of using the object oriented interface. It creates a figure (fig) that can hold many different elements, allowing for greater customization.

For each line of code, describe what the line does. If there are multiple features in a line of code, describe each element. Again, feel free to manipulate variables to aid in your exploration.

```python
fig, ax = plt.subplots(2, 2, figsize=(10,5)) # Subplots indicates multiple graphs in one spread 
# The first two numbers indicate the number of graphs; the 1st number is # of rows, 2nd is # of columns 
# figsize indicates size of figures 
fig.subplots_adjust(hspace=0.5) # This adjusts the space in between the first and second row of graphs 

# All graphs are dictated in position by the 'ax[#,#]' function. The 1st number denotes top(0) and bottom (1) 
# and the 2nd number denotes left(0) and right (1) 

# The following code plots two lines on the top left graph, each with a different color and equation 
# The first position is the x-axis values while the second is the y-values, which is why the modified equation is 
# changed in the second position from yvals to 'xvals*2+4' 
ax[0,0].plot(xvals, yvals, color= 'chartreuse', label='one', linestyle="--") # This code colors green, labels the line, and dots the line 
ax[0,0].plot(xvals, xvals*2+4, color= 'skyblue', label='two', linestyle="--") # This code colors blue, labels line, and makes it dotted 
ax[0,0].legend() # This takes the pre-made labels into a legend in the top left corner 

# This is for the bottom left graph 
ax[1,0].set_title("this is a title") # This adds a title 
ax[1,0].annotate('there is nothing here',xy=(.5, .5), # This identifies where the point of interest for the annotation takes place 
                 xytext=(.2, .7), # This creates a line of text inserted with the 'xytext' customization 
                 arrowprops=dict(color='black', arrowstyle="->")) # This creates an arrow that is black and a ertain style (arrowstyle) 
# The following section seems like it should create a visible line bordering the top, that is black (I changed the hexcode myself) 
ax[1,0].spines['top'].set_visible(True) # The code *should* show a solid black line running across the top x-axis 
ax[1,0].spines['right'].set_visible(True) # The code *should* show a solid black line running across the right y-axis 
# The two above also use a boolean for whether or not the spine shows! 
ax[1,0].spines['bottom'].set_color('#000000') # This sets the color of the bottom axis with a hex code 

# The code for the top right shows an empty, but otherwise standard, plot as we did previously in labs/code 
ax[0,1].set(xlabel="some label", ylabel="other label", title="a title?") # This labels both axes and adds a title 
ax[0,1].set_xlim(2,5) # This sets the range of the x-axis from 2 to 5, and although inputs are integers, the axis is in floats 
ax[0,1].xaxis.set_ticks_position('bottom') # This determines where labels for the x-axis values are placed. 'none' defaults to 'bottom' 
ax[0,1].tick_params(labelcolor='navy') # This assigns a color to the parameter values along both axes. I changed it to navy. 
ax[0,1].grid(color='#FF69B4', axis='y', linestyle='-', linewidth=0.25, alpha=1.0) # This sets the color of the y-axis grid lines 
                                                                                # The code also determines the line type (dotted, for example) 
                                                                                # and the width + opacity 
# The grid line code can also be applied to the x accordingly: 
ax[0,1].grid(color='#FF69B4', axis='x', linestyle='--', linewidth=0.25, alpha=1.0) # I made the x-axis lines dotted this time 

# This last graph is the bottom right, and doesn't have a title or axes labels 
ax[1,1].axhline(.5, color='purple') # This creates one solid colored line, at the y-axis position indicated by the 1st number 
ax[1,1].text(.7, .7, "text!") # This indicates the position and contents of text of the graph 

plt.show() # This last line is the equivalent of 'print()' for all subplots 
```

```python
# Note: any customizations not already included in the original code, I got from the cheatsheets or 
# Matplotlib's documentation page: https://matplotlib.org/stable/index.html 

# For ease of navigation, I removed comments from the following code that are present in the first 
fig, ax = plt.subplots(2, 2, figsize=(10,5)) # Subplots indicates multiple graphs in one spread 
fig.subplots_adjust(hspace=0.5) # This adjusts the space in between the first and second row of graphs 

# Top left 
ax[0,0].plot(xvals, yvals, color= 'chartreuse', label='one', linestyle="--") # Original code 
ax[0,0].plot(xvals, xvals*2+4, color= 'skyblue', label='two', linestyle="--") # OG code 
ax[0,0].set_title("A title") # Added a title 
ax[0,0].set(xlabel = "Some label", ylabel = "Another label") # Added axes label 

# Using the numpy.polyfit function, I fited a 2 degree polynomial to the points (xvals, yvals) 
# I then plotted this to confirm, and used Mathematica to find the points which I stored in (x1, y1) 
# The Mathematica code I used was: Solve[2x^2-7 == 2x+4, x]
x1 = float((1/2)*(1+np.sqrt(23)))
y1 = float(2*x1+4) 

ax[0,0].scatter(x1, y1) # This plots the intersection. The code to confirm the intersection is in the 'Additional Notes' 
ax[0,0].annotate('A little note', # Change the text of the annotation to match your maker 
                 xy=(x1, y1), # This is where the arrow points 
                 xytext=(2, 18), # This is where the text starts 
                 arrowprops=dict(color='black', arrowstyle="->")) # Change the style of the arrow to fancy 
ax[0,0].legend(bbox_to_anchor=(0.95, 0.4), # Changed the location of the legend 
                         loc='upper right', borderaxespad=0.) 

# Bottom left 
ax[1,0].set_title("This is a title") 
ax[1,0].annotate('*', # Changed the text of the annotation to match my marker 
                 xy=(.5, .5), # This is where the arrow points Change the text of the annotation to match your maker 
                 xytext=(.2, .7), # This is where the text starts 
                 arrowprops=dict(color='black', arrowstyle="fancy")) # Change the style of the arrow to fancy 
ax[1,0].scatter(.5,.5, marker = '*') # This plots the a marker at the tip of the arrow 
ax[1,0].set_xlim(0,1) # Limits x-axis to prevent crazy scale 
ax[1,0].set_ylim(0,1) # Same thing for y-axis too 
ax[1,0].spines['top'].set_visible(True) # Ensure the top axis is visible 
ax[1,0].spines['right'].set_visible(True) # Same thing for right 
ax[1,0].spines['bottom'].set_color('#000000') # Same thing for bottom, and add a hexcode 

# Top right 
ax[0,1].set(xlabel="Some label", ylabel="Other label", title="A title?") 
ax[0,1].scatter(xvals, yvals, marker = '*') # Add a scatter plot of yvals versus xvals using a marker of your choice 

# Notice how we lose 4 data points, since the log scale can't accommodate non-positive values (including zero) 
ax[0,1].set_xlim(2,5.2) # Sets the limit for the x-axis, which is unaffected by the log scale but should account for lost negative values 
ax[0,1].set_yscale("log", base=10) # This creates a log scale 
ax[0,1].set_ylim(1,100) # Fits the graph to the new scale 
ax[0,1].xaxis.set_ticks_position('bottom') # This ensures the x-axis parameter values run along the bottom 
ax[0,1].tick_params(labelcolor='navy') # Adds the color of the parameter values 
ax[0,1].grid(color='#FF69B4', axis='x', linestyle='--', linewidth=0.25, alpha=1.0) # Add vertical grid lines in pink 

# Bottom right 
# In the following curve, I try to model a Keeling curve, which is infamous for demonstrating the amount of carbon dioxide 
# in the atmosphere. Although, the real Keeling curve has a slight exponential increase, while mine is largely linear aside 
# from the oscillations from the trig function. 
years = np.linspace(2000, 2030, num=2000) # The number of values stored in the array has to be high to create a smooth-appearing function 
carbonEmissions = np.sin(5*years)+(2/9)*years # This is a trig function added to a linear 'x' so it increases like the Keeling curve 

ax[1,1].plot(years, carbonEmissions, alpha = 0.5) # This produces a simple plot with my 'Keeling curve'; slightly opaque in color 
# ax[1,1].set_xlim(2000,2030) # In this case, the axis works better than setting a solid limit, so I commented this out 

# The following lines sets the titles and axes labels 
ax[1,1].set_title("Carbon Emissions since 2000")
ax[1,1].set_xlabel("Year") 
ax[1,1].set_ylabel("CO2 Concentration (ppm)", fontsize = 9)
ax[1,1].grid('True') # Adds a grid for reference 

# In the next section, I will make an annotation to where our present year is. Note, this data is not real! 
def co2Emissions(x): # Here, I am essentially defining an actual function to identify the actual point with x = 2025 
    y = np.sin(5*x)+(2/9)*x
    return float(y) # This ensures a float data type object is given! I found the print() command to not work in that respect  

# This is copied from the standard annotation format of the top left graph! 
ax[1,1].annotate(text = 'Today', # Cause right now we are in 2025 
                 xy = (2025, co2Emissions(2025)), # This is where the arrow will point 
                 xytext=(2025,448), # This is where the text starts 
                 xycoords='data', # Ensures the coordinate axes are the same for the arrow and data 
                 textcoords='data', # Same, but for the text annotation 
                arrowprops=dict(color='black', arrowstyle="->")) # Creates the arrow 

plt.show() # This last line is the equivalent of 'print()' for all subplots 
```
